Iris Flower classification using decision tree

In this task we will create a decision tree for iris data set and visualize it graphically.

We will start by importing the data and creating data frame.

```python
import pandas as pd
from sklearn import tree
import numpy as np
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width','flower']
df = pd.read_csv('iris.data', names=column_names)
df.head()
```

Our Decision Tree model requires the data in numerical form so we need to convert the flower names to numbers 

iris-setosa = 0, iris-versicolor = 1, iris-virginica = 2

```python
d = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
df['flower'] = df['flower'].map(d)
df.head()
```

Now, we will create list of feature columns of sepal and petal lengths and widths

```python
features = list(df.columns[:4])
features
```

Creating X and y data for training. X will contain or features and y will contain the flower types

```python
y = df['flower']
X = df[features]
```

Now we will construct the decision tree with the help of decision tree classifier from sklearn

```python
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(X,y)
```

We will display our Decision tree with the help of sklearn

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,16))
tree.plot_tree(classifier, feature_names= features, class_names=class_names, filled=True)
```

We have successfully build decision tree and visualized it very easily.

We will now give any random features to the classifier and see which class it belongs to

```python
test_features = [7,6,3,2]
classifier.predict([test_features])
```

Here, we can see that above features belong to class 1 i.e. Iris-versicolor flower
